```python
import os

import xarray as xr
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
folder = os.getcwd()

fp = os.path.join(folder, "xboutput.nc")

ds = xr.open_dataset(fp)
```

```python
print(ds)
```

```python
ds = ds.squeeze(drop=True)
```

```python
print(ds)
```

```python
x = ds.globalx.values
z = ds.zb.values

wl = ds.zs.values

sedero = ds.sedero.values
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(x, z, label='bathymetry')
ax.plot(x, wl, label='water level')
ax.plot(x, sedero, label='sedero')

ax.set_xlabel('x [m]')

# ax.set_aspect('equal')

ax.legend()
```
