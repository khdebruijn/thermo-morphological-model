```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
import xarray as xr
```

# Start with Shoreline and Bluff edge

```python
# Read the Shapefile into a GeoDataFrame
gdf_be = gpd.read_file('gibbs/BarterIsland_BluffEdge_1950to2020/BarterIsland_BluffEdge_1950to2020.shp')
gdf_sl = gpd.read_file('gibbs/BarterIsland_Shoreline_1947to2020/BarterIsland_Shoreline_1947to2020.shp')
```

```python
gdf_be.describe()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

gdf_be.plot(ax=ax, color='C0', alpha=0.2)
gdf_sl.plot(ax=ax, color='C1', alpha=0.2)
```

# Secondly, Shoreline and Bluff edge change rate

```python
# Read the Shapefile into a GeoDataFrame
gdf_be_cr = gpd.read_file('gibbs/BarterIsland_BluffTransect_ChangeRates_1950to2020/BarterIsland_BluffTransect_ChangeRates_1950to2020.shp')
gdf_sl_cr = gpd.read_file('gibbs/BarterIsland_ShoreTransect_ChangeRates_1947to2020/BarterIsland_ShoreTransect_ChangeRates_1947to2020.shp')
```

```python
gdf_be_cr.describe()
```

```python
len(gdf_be_cr.TransectID.unique())
```

```python
gdf_sl_cr.describe()
```

```python
len(gdf_sl_cr.TransectID.unique())
```

```python
transect_id = 300

fig, ax = plt.subplots(figsize=(15, 5))

gdf_be.plot(ax=ax, color='C0', alpha=0.2)
gdf_sl.plot(ax=ax, color='C1', alpha=0.2)

gdf_be_cr[gdf_be_cr.TransectID==transect_id].plot(ax=ax, color='red')
gdf_sl_cr[gdf_sl_cr.TransectID==transect_id].plot(ax=ax, color='red')
```
