```python
df_be = gpd.read_file(Path("ts_datasets/bluff_edges.csv"))
sort_mask = np.argsort(df_be.time.values)
df_be = df_be.iloc[sort_mask]

df_be['time'] = pd.to_datetime(df_be.time.values)
df_be['distance_from_ref'] = np.float32(df_be['distance_from_ref'].values)

df_sl = gpd.read_file(Path("ts_datasets/shore_lines.csv"))
sort_mask = np.argsort(df_sl.time.values)
df_sl = df_sl.iloc[sort_mask]

df_sl['time'] = pd.to_datetime(df_sl.time.values)
df_sl['distance_from_ref'] = np.float32(df_sl['distance_from_ref'].values)


df_be.head()
```

```python
def get_rates_and_plot_dates(df):

    def difference_in_years(date1, date2):
        
        date1 = pd.to_datetime(date1)
        date2 = pd.to_datetime(date2)
        
        # difference_in_days = (date2_dt - date1_dt).days
        
        difference_in_days = (date2 - date1).days

        difference_in_years = difference_in_days / 365.25
        
        return difference_in_years

    dx = df['distance_from_ref'].values[1:] - df['distance_from_ref'].values[:-1]
    dt = np.zeros(dx.shape)

    for i, date1, date2 in zip(np.arange(len(df)), df.time.values[:-1], df.time.values[1:]):
        
        dt[i] = difference_in_years(date1, date2)
    
    dx_dt2 = np.repeat(dx/dt, 2)

    t2 = np.repeat(df.time.values, 2)
    
    t2[1::2] += pd.to_timedelta(1, 's')
    
    t2 = t2[1:-1]
    
    return (t2, dx_dt2)

be_t2, be_dx_dt2 = get_rates_and_plot_dates(df_be)
sl_t2, sl_dx_dt2 = get_rates_and_plot_dates(df_sl)
```

```python
fig, ax = plt.subplots(figsize=(15,6))

ax.fill_between([pd.Timestamp('2009-08-02'), pd.Timestamp('2011-07-15')], 0, 200, alpha=0.3, label='low erosion', color='C2')
ax.fill_between([pd.Timestamp('2012-07-11'), pd.Timestamp('2015-07-05')], 0, 200, alpha=0.3, label='high erosion', color='C3')
ax.fill_between([pd.Timestamp('2016-08-27'), pd.Timestamp('2018-07-30')], 0, 200, alpha=0.3, label='high and low erosion', color='C4')


ax.plot(df_be.time.values, df_be.distance_from_ref.values, label='bluff edge', color='C0')
ax.plot(df_sl.time.values, df_sl.distance_from_ref.values, label='shore line', color='C1')

ax.scatter(df_be.time.values, df_be.distance_from_ref.values, color='C0')
ax.scatter(df_sl.time.values, df_sl.distance_from_ref.values, color='C1')

ax.set_xlim((pd.Timestamp('2000-01-01'), pd.Timestamp('2020-01-01')))
ax.set_ylim((120, 200))

ax.grid()

ax.set_xlabel('time')
ax.set_ylabel('distance from reference [m]')

ax.set_title('Historic shore line and bluff edge location, with validation periods')

ax.legend(loc='upper left')

fig.tight_layout()

fig.show()

figname = 'model_validation_periods_a.png'

fig.savefig(os.path.join(save_path, figname), bbox_inches='tight', facecolor='white', transparent=False)
```
